# FBREF

https://fbref.com/ is a popular website for more advanced statistics. The nice thing about FBREF is that everything is just in tables so we can scrape it really easily just using pandas.

So while we'll only be scraping just a single table in this video, this method is applicable to any table on the website.

```python
#!pip install undetected-chromedriver
# make sure you also have Chrome installed

import pandas as pd
import undetected_chromedriver as uc
```

```python
driver = uc.Chrome()
driver.set_page_load_timeout(5)
```

```python
driver.get('https://fbref.com/en/comps/8/2023-2024/stats/2023-2024-Champions-League-Stats')
```

```python
# We'll start off by scraping the data from the 23/24 Champions League
df = pd.read_html(driver.page_source, attrs={'id': 'stats_squads_standard_for'})[0]
```

```python
df.head()
```

```python
# You'll also notice that the columns are multi-indexed, so we'll flatten the columns
df.columns = ['_'.join(col).strip() for col in df.columns.values]
```

```python
df.head()
```

```python
# You can remove some of the names if you want
df.columns = df.columns.str.replace('Unnamed: 0_level_0_', '')
df.columns = df.columns.str.replace('Unnamed: 1_level_0_', '')
df.columns = df.columns.str.replace('Unnamed: 2_level_0_', '')
df.columns = df.columns.str.replace('Unnamed: 3_level_0_', '')
```
